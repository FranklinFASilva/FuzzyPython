Importando as bibliotecas

```
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

Criando as variáveis

```
frequencia = ctrl.Antecedent(np.arange(0, 8, 1), 'frequencia')
tempo = ctrl.Antecedent(np.arange(20, 91, 1), 'tempo')
peso = ctrl.Antecedent(np.arange(0, 41, 1), 'peso')
adesao = ctrl.Antecedent(np.arange(0, 7, 1), 'adesao')
condicionamento = ctrl.Antecedent(np.arange(16, 81, 1), 'condicionamento')
intensidade = ctrl.Consequent(np.arange(40, 101, 1), 'intensidade')
```

Criando as funções de pertinência

```
#função para frequência
frequencia['baixa'] = fuzz.trapmf(frequencia.universe, [0, 0, 2, 4])
frequencia['média'] = fuzz.trimf(frequencia.universe, [2, 4, 6])
frequencia['alta'] = fuzz.trapmf(frequencia.universe, [4, 6, 7, 7])

#função para tempo disponível
tempo['pouco'] = fuzz.trapmf(tempo.universe, [20, 20, 30, 40])
tempo['médio'] = fuzz.trimf(tempo.universe, [30, 40, 50])
tempo['muito'] = fuzz.trapmf(tempo.universe, [40, 60, 90, 90])

#função para peso
peso['pouco'] = fuzz.trapmf(peso.universe, [0, 0, 5, 15])
peso['médio'] = fuzz.trimf(peso.universe, [5, 15, 30])
peso['muito'] = fuzz.trapmf(peso.universe, [15, 30, 40, 40])

#função para adesão
adesao['baixa'] = fuzz.trimf(adesao.universe, [0, 0, 3.5])
adesao['média'] = fuzz.trimf(adesao.universe, [0, 3.5, 7])
adesao['alta'] = fuzz.trimf(adesao.universe, [3.5, 7, 7])

#função para condicionamento
condicionamento['baixo'] = fuzz.trapmf(condicionamento.universe, [16, 16, 20.5, 29.5])
condicionamento['baixo/médio'] = fuzz.trimf(condicionamento.universe, [20.5, 29.5, 38.5])
condicionamento['médio'] = fuzz.trimf(condicionamento.universe, [29.5, 38.5, 47.5])
condicionamento['médio/alto'] = fuzz.trimf(condicionamento.universe, [38.5, 47.5, 56.5])
condicionamento['alto'] = fuzz.trapmf(condicionamento.universe, [47.5, 56.5, 80, 80])

#função para intensidade
intensidade['baixa'] = fuzz.trapmf(intensidade.universe, [40, 40, 47.5, 57.5])
intensidade['baixa/média'] = fuzz.trimf(intensidade.universe, [47.5, 57.5, 72.5])
intensidade['média'] = fuzz.trimf(intensidade.universe, [57.5, 72.5, 82.5])
intensidade['média/alta'] = fuzz.trimf(intensidade.universe, [72.5, 82.5, 95])
intensidade['alta'] = fuzz.trapmf(intensidade.universe, [82.5, 95, 100, 100])
```

Criando as regras

```
rule1 = ctrl.Rule(condicionamento['alto'], intensidade['alta'])
rule2 = ctrl.Rule(condicionamento['médio/alto'], intensidade['média/alta'])
rule3 = ctrl.Rule(condicionamento['médio'], intensidade['média'])
rule4 = ctrl.Rule(condicionamento['baixo/médio'], intensidade['baixa/média'])
rule5 = ctrl.Rule(condicionamento['baixo'], intensidade['baixa'])
rule6 = ctrl.Rule(tempo['muito'] & condicionamento['alto'], intensidade['média/alta'])
rule7 = ctrl.Rule(tempo['muito'] & condicionamento['médio/alto'], intensidade['média'])
rule8 = ctrl.Rule(tempo['muito'] & condicionamento['médio'], intensidade['baixa/média'])
rule9 = ctrl.Rule(frequencia['baixa'] & condicionamento['alto'], intensidade['alta'])
rule10 = ctrl.Rule(frequencia['baixa'] & condicionamento['médio/alto'], intensidade['alta'])
rule11 = ctrl.Rule(frequencia['baixa'] & condicionamento['médio'], intensidade['média/alta'])
rule12 = ctrl.Rule(frequencia['baixa'] & condicionamento['baixo/médio'], intensidade['média'])
rule13 = ctrl.Rule(frequencia['baixa'] & condicionamento['baixo'], intensidade['baixa/média'])
rule14 = ctrl.Rule(tempo['pouco'] & condicionamento['alto'], intensidade['alta'])
rule15 = ctrl.Rule(tempo['pouco'] & condicionamento['médio/alto'], intensidade['alta'])
rule16 = ctrl.Rule(tempo['pouco'] & condicionamento['médio'], intensidade['média/alta'])
rule17 = ctrl.Rule(tempo['pouco'] & condicionamento['baixo/médio'], intensidade['média'])
rule18 = ctrl.Rule(tempo['pouco'] & condicionamento['baixo'], intensidade['baixa/média'])
rule19 = ctrl.Rule(frequencia['alta'] & condicionamento['alto'], intensidade['média/alta'])
rule20 = ctrl.Rule(frequencia['alta'] & condicionamento['médio/alto'], intensidade['média'])
rule21 = ctrl.Rule(frequencia['alta'] & condicionamento['médio'], intensidade['baixa/média'])
rule22 = ctrl.Rule(frequencia['alta'] & condicionamento['baixo/médio'], intensidade['baixa'])
rule23 = ctrl.Rule(adesao['alta'] & condicionamento['alto'], intensidade['alta'])
rule24 = ctrl.Rule(adesao['alta'] & condicionamento['médio/alto'], intensidade['alta'])
rule25 = ctrl.Rule(adesao['alta'] & condicionamento['médio'], intensidade['média/alta'])
rule26 = ctrl.Rule(adesao['alta'] & condicionamento['baixo/médio'], intensidade['média'])
rule27 = ctrl.Rule(adesao['alta'] & condicionamento['baixo'], intensidade['baixa/média'])
rule28 = ctrl.Rule(adesao['baixa'] & condicionamento['alto'], intensidade['média/alta'])
rule29 = ctrl.Rule(adesao['baixa'] & condicionamento['médio/alto'], intensidade['média'])
rule30 = ctrl.Rule(adesao['baixa'] & condicionamento['médio'], intensidade['baixa/média'])
rule31 = ctrl.Rule(adesao['baixa'] & condicionamento['baixo/médio'], intensidade['baixa'])
rule32 = ctrl.Rule(adesao['média'] & condicionamento['alto'], intensidade['alta'])
rule33 = ctrl.Rule(adesao['média'] & condicionamento['médio/alto'], intensidade['média/alta'])
rule34 = ctrl.Rule(adesao['média'] & condicionamento['médio'], intensidade['média'])
rule35 = ctrl.Rule(adesao['média'] & condicionamento['baixo/médio'], intensidade['baixa/média'])
rule36 = ctrl.Rule(adesao['média'] & condicionamento['baixo'], intensidade['baixa'])
rule37 = ctrl.Rule(peso['médio'] & condicionamento['baixo'], intensidade['baixa/média'])
rule38 = ctrl.Rule(peso['médio'] & condicionamento['baixo/médio'], intensidade['média'])
rule39 = ctrl.Rule(peso['médio'] & condicionamento['médio'], intensidade['média/alta'])
rule40 = ctrl.Rule(peso['médio'] & condicionamento['médio/alto'], intensidade['alta'])
```

Simulando o controlador Fuzzy

```
intensidade_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27, rule28, rule29, rule30, rule31, rule32, rule33, rule34, rule35, rule36, rule37, rule38, rule39, rule40])
intensidade_simulador = ctrl.ControlSystemSimulation(intensidade_ctrl)
```

```
#criando valores e simulando o resultado
intensidade_simulador.input['frequencia'] = 6
intensidade_simulador.input['tempo'] = 20
intensidade_simulador.input['peso'] = 3
intensidade_simulador.input['adesao'] = 3
intensidade_simulador.input['condicionamento'] = 47
```

Calculando a intensidade

```
intensidade_simulador.compute()
```

Imprimindo

```
print("A intensidade aplicada deve ser de: {:.2f}".format(intensidade_simulador.output['intensidade']),"%")
```
